##202044072-2 신도림역의 두 호선 중 어디가 더 많이 이용될까?

###전처리

```python
sl=s.groupby(['호선명']).mean().reset_index()
sl#확인
```

###평균 승차 인원 수 

```python
on=pd.DataFrame()
on['호선명']=sl['호선명']
for i in range(int((len(sl.columns)-2)/2)):
  on[sl.columns[2+2*i]]=sl[sl.columns[2+2*i]]
on=on.set_index('호선명')
on
plt.figure(figsize=(20,10))
on.mean().plot(kind=('bar'))
plt.show()
```

###평균 하차 인원수

```python
off=pd.DataFrame()
off['호선명']=sl['호선명']
for i in range(int((len(sl.columns)-2)/2)):
  off[sl.columns[3+2*i]]=sl[sl.columns[3+2*i]]
off_s=off.set_index('호선명')
off
plt.figure(figsize=(20,10))
off_s.mean().plot(kind=('bar'))
plt.show()
```

###호선 별 평균 승하차 인원 구하기

```python
hosun=pd.DataFrame(index=sl['호선명'])
hosun['평균 승차 인원 수']=on.mean(axis=1)
hosun['평균 하차 인원 수']=on.mean(axis=1)
hosun
```

###데이터 시각화

```python
plt.figure(figsize=(20,10))
hosun.plot(kind=('bar'))
plt.show()
```

**2호선이 승차 인원,하차 인원 모두 더 많다**

##년도별 신도림역 평균 이용

```python
year=pd.DataFrame({'년도':[2015,2016,2017,2018,2019,2020,2021,2022]
                   ,'이용 평균':[five.mean().mean(),six.mean().mean(),
                             seven.mean().mean(),eight.mean().mean(),
                             nine.mean().mean(),two.mean().mean(),
                             one.mean().mean(),tt.mean().mean()],
                   '승차 평균':[five_on.mean().mean(),six_on.mean().mean(),
                             seven_on.mean().mean(),eight_on.mean().mean(),
                             nine_on.mean().mean(),two_on.mean().mean(),
                             one_on.mean().mean(),tt_on.mean().mean()],
                   '하차 평균':[five_off.mean().mean(),six_off.mean().mean(),
                             seven_off.mean().mean(),eight_off.mean().mean(),
                             nine_off.mean().mean(),two_off.mean().mean(),
                             one_off.mean().mean(),tt_off.mean().mean()]
                   })
year=year.set_index('년도')
year
```

###코로나로 얼마나 줄었을까?
